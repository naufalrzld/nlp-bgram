# Importing library and dataset

```python
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
import os
import sys
import json
```

```python
dataset = pd.read_csv('plain/PANL-BPPT-SCI-ID-100Kw.txt', sep='\t', header=None)
dataset = np.array(dataset)

temp = ''
for doc in dataset:
    temp += doc
    
dataset = temp
del temp
```

```python
print(dataset)
len(dataset[0])
```

```python
# Make a unigram model

token = word_tokenize(dataset[0].lower())

corpus = []
for word in token:
    if word not in corpus:
        corpus.append(word)
```

```python
print(corpus)
len(corpus)
```

```python
print(token)
```

## Bigram Model

```python
# Make a bigram model

dictio = {}
i = 0
for kata in corpus:
    print(i+1,'/',len(corpus))
    temp = {}
    for item in corpus:
        temp[item] = 0
    dictio[kata] = temp
    i+=1
```

```python
# corpus = is all unique word in dataset

print('Banyak kata unik: ',len(corpus))
```

```python
y = np.array(corpus)
token = np.array(token)
i = 0
for item in y:
    print(i,'/',len(y))

    # find index for each word
    find = np.where(token == item)
    find_after = find[0]+1
    print('item: ', item)
    print('find: ', find[0])
    print('find after: ', find_after)

    # catch error if index in the last potition 
    check_last = find_after[len(find_after)-1:]
    print('check last', check_last)
    
    # if word in last sentence, so we don't counting in
    if check_last[0] == len(token):
        find_after = find_after[:-1]

    array = token[find_after]
    print('find after', find_after)
    print('array', array)

    for kata in array:
        dictio[item][kata]+=1
        
    i+=1
```

```python
dictio['zahir']['zahir'] = 0
```

```python
# save bigram model

with open('bigram.json', 'w') as fp:
    json.dump(dictio, fp)
```

# Make a probability model of language model

```python
# Load unigram model

with open('unigram.json', 'r') as f:
    unigram = json.load(f)
```

```python
unigram
```

```python
# Language Modelling model

i=0
for item in corpus:
    print(i,'/',len(corpus))
    i+=1
    for kata in corpus:
        if dictio[item][kata] != 0:
            dictio[item][kata] = dictio[item][kata]/unigram[item]
```

```python
# save bigram model

with open('language_model.json', 'w') as fp:
    json.dump(dictio, fp)
```
